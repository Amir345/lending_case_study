### Since our dataset is clean now and there are no missing values, let's add some derived columns that may be helpful for analysis
* bucket of emp length (Associate(<3), Senior(<=6), Specialist(>6))

```python
##Checking distribution of annual income
sns.boxplot(bank_data.annual_inc)
plt.show()
```

Since annual income has **outliers** let's remove these outliers and assign it to a temp variable

```python
##Based on P99 removing the outlier
tmp=bank_data[bank_data["annual_inc"] < bank_data["annual_inc"].quantile(0.99)]
```

```python
## plotting the box plot of the temp dataset and annual income
sns.boxplot(tmp.annual_inc)
plt.show()
```

```python
## looking at the annual income column of the temp dataset that was created earlier
tmp.annual_inc.describe(percentiles=[0.90])
```

```python
##checking the mean for annual_inc column with original dataset
bank_data.annual_inc.mean()
```

Since after removing the outlier our mean has not changed a lot, let's use this tmp variable to create buckets for annual inc

```python
# Creating buckets for annual incomes 
bank_data['annual_inc_bucket'] = pd.cut(tmp['annual_inc'], [0, 20000, 40000, 60000, 80000,1000000], labels=['lt_20000', '20000-40000', '40000-60000', '60000-80000', 'gt_80000'])
```

```python
###Checking the distribution of loan_amount
sns.boxplot(bank_data.loan_amnt)
plt.show()
```

Since the upper fence is at **30K**, let's cut it in **6K** each

```python
#Creating buckets for  loan amounts 
bank_data['loan_amnt_bucket'] = pd.cut(bank_data['loan_amnt'], [0, 6000, 12000, 18000, 24000, 30000], labels=['lt_6000', '6000-12000', '12000-18000', '18000-24000', 'gt_24000'])
```

```python
##looking at the buckets of loan amount
bank_data.loan_amnt_bucket.value_counts()
```

```python
## Extracting emp_length in year and keeping the original column as well
bank_data['emp_length_years']=bank_data['emp_length'].apply(lambda x : ("".join(filter(str.isdigit, str(x))))).values
```

```python
bank_data.head()
```

```python
## Converting the datatype of the newly generated column as integer
bank_data.emp_length_years=bank_data.emp_length_years.astype('int64')
```

```python
## Creating buckets for Experience based on emp_length_years
bank_data['Experience']=bank_data.emp_length_years.apply(lambda x: 'Associate' if x <=3 else 'Senior' if (x>3 and x<=6) else 'Specialist')
```

```python
bank_data.head()
```

```python
## creating a boolean column assigning "True" for charged off and "False" for Fully Paid
bank_data['loan_status_bool']=bank_data.loan_status.apply(lambda x: False if x == 'Fully Paid' else True )
```

```python
## plotting to check th IQR range for dti
plt.figure(figsize=(10,5))
bank_data.dti.plot.box()
plt.show()
```

* Based on the above plot creating three buckets for dti

```python
## Creating buckets for dti
bank_data['dti_range']=bank_data.dti.apply(lambda x: 'Low' if x <=10 else 'Med' if (x>10 and x<=20) else 'High')
```

```python
bank_data.head()
```

```python
## checking datatype for issue date
bank_data.issue_d.dtype
```

* We can see that the datatype is **object** hence let's take out year and month from the column

```python
## looking at unique values and their count for the issue_d column
bank_data.issue_d.value_counts()
```

```python
##extracting year from the issue_d column
bank_data['loan_year']=bank_data.issue_d.apply(lambda x: x.split('-')[1])
```

```python
# extracting month from the issue_d column
bank_data['loan_month']=bank_data.issue_d.apply(lambda x: x.split('-')[0])
```

# Insights from the above analysis

* Loan Default rate is 14.5%, i.e out of every 100 loans issues 14 are defaulting. Default loans has costed the company a net loss of **~426 million** dollars which is equal to 16% of total loan amount issued


* Lending club should scrutinize more on loan amount greater than 20K, as the probability of default increases.


* Lending club can improve on their interest rates on loan amount, as we have seen more the interest rate more the chances of default


* Lending club should bring more guardrails for loan category of B,C,D as they are defaulting more in number, low grade loans are riskier hence we should cut down the acceptance rate of low grade loans (E,F,G). Within B,C,D category following sub grades should be more focused (B5,B3,C1,B4,C2) and for low grade loans following sub grades should be put under more checks(F5,G3,G5,G2,F4). E,F,G grades are asking for more loan amounts compare to other grades. Also the loan default rate is higher in E,F,G hence more scrutiny should be done for these loan grades before approving loans


* Lending club should pay more attention to employees who has experience of more than 10 yeas or less than 3 years as they are defaultig more. Associates are asking for amount majorly in around 0-12K range, hence we should focus more on employees greater than 7 years of experience as they are taking more money and defaulting more.


* People who are living on Rent are defaulting more, hence additional control should be in place for people who are living on a rented property. Maybe cutdown their requested loan amount.


* Overall, it appears that there are more 'Non Verified' loans which are leading to charge off loans. LC should take steps to verify these loans so as to reduce the Charge Off percentage. Though there are charge off which are largely seen in verified loans too.


* Default rate is seen higher than average in (Sep-Dec) Winter holidays, Lending club should be cautious of the loan application they get during this period. December should be more focused as loan amount requested has increased and the default rate as well.


* Lending Club should be wary of the loans where the purpose is Small Business as the percentage of a loan being charged off is highest, Apart from small business LC should also focus on categories like debt consolidation and renewable energy. Small business category is also asking for more loan amount with debt consolidation following up in terms of loan amount.


* LC should put an additional check on states like (Nebraska > Nevada > Alaska) as the default rate is higher and states like (California > Florida > NewYork > Texas) as the number of default count is higher. People from state Nebraska and Alaska are asking for more loan amount as compared to other states and also defaulting more so these two states should be our focus


* LC should check for loan term of 5 years as they are defaulting more. Also people who opt for 5 years are requesting more loan amount resulting an increase in loss (e.g 3, 10K loans with term of 3 years get paid fully and 1, 5 year term loan asking 30K default resulting in a big loss for the company)


* Loan default rate is higher in income group where the income is less than 20k. Count of default loans are in order (40K-60k>20K-40K>gt_80K), hence focus should be more for income group 20K to 40K


* LC should check on Debt to income, as when DTI is less people are defaulting less and default rate rises in relation to increase in DTI


* People who has a low DTI (less than 10) has high annual income hence the default rate is also less. We can say that people with higher income are less likely to "Default". Lending club should do through income checks before approving loans.


* Category Debt Consolidation is asking for more time to pay, and as we know that more the time duration to pay loan the chances of default increases, hence lending club should put more guardrail on loans with debt_consolidation category
